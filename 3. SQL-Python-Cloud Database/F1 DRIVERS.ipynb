### 2. Demonstrate core SQL statements: SELECT / DISTINCT / COUNT / LIMIT / ORDER BY / GROUP BY / AND 

We use SELECT * to view drivers.csv in full, with a LIMIT of the first 20 rows of data. 

```python
%sql SELECT * from DRIVERS LIMIT 20
```

We count the total number of rows in this dataset.

```python
%sql SELECT COUNT(*) from DRIVERS
```

We are interested in the data of driver ID 102. We choose/rearrange columns to be displayed, based on our requirements.  

```python
%%sql SELECT forename, surname, nationality, dob, driverid 
from DRIVERS where driverid='102'
```

What unique (non-duplicated) last names do German drivers have?

```python
%sql SELECT DISTINCT surname from DRIVERS where nationality='German'
```

We COUNT the number of drivers by nationality. Then, we display the top 10 results in descending order. 

```python
%%sql SELECT nationality, COUNT(*) as number_of_drivers 
from DRIVERS GROUP BY nationality ORDER BY number_of_drivers desc LIMIT 10
```

We convert the above table into a dataframe structure, then we use Matplotlib to generate a horizontal bar graph.

```python
#SQL variable assignment.
top10 = %sql SELECT nationality, COUNT(*) as number_of_drivers from DRIVERS GROUP BY nationality ORDER BY number_of_drivers desc LIMIT 10

# We construct a pandas dataframe using top10 as data source.
df = pd.DataFrame(top10,columns=['Nationalities','Number of Drivers'])

df.plot(kind='barh', x='Nationalities', y='Number of Drivers')

#adding a proper title
plt.title("Top 10 Nationalities of F1 Drivers")
```

Here are all the F1 drivers of Canadian nationality, ordered by first names in ascending order. 

```python
%%sql SELECT forename, surname, nationality, dob
from DRIVERS where nationality='Canadian' ORDER BY forename asc
```

Here are the 10 youngest F1 drivers. 

```python
%sql SELECT forename, surname, dob, url from DRIVERS ORDER BY dob desc LIMIT 10
```

Here are all the F1 drivers who are officially assigned a car number. We do this by negating (not) any rows with '\N' under the number column. 

```python
%%sql SELECT forename, surname, nationality, number
from DRIVERS where not number='\N' 
ORDER BY forename
```

With the same list, we further narrow down our criteria by specifying nationality to be 'German' and surname to be 'Schumacher'.

```python
%%sql SELECT forename, surname, dob, nationality, number, url
from DRIVERS where not number='\N' and nationality='German' and surname='Schumacher'
ORDER BY forename
```
